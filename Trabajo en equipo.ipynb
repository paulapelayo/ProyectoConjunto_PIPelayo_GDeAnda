```python
# Program to display the Fibonacci sequence up to n-th term

nterms = int(input("How many terms? "))

# first two terms
n1, n2 = 0, 1
count = 0

# check if the number of terms is valid
if nterms <= 0:
   print("Please enter a positive integer")
# if there is only one term, return n1
elif nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)
# generate fibonacci sequence
else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```

```python
#Introducción al programa de números primos
print("Bienvenido a tu programa de tus números primos con un rango del 1 al 100")
print("Cargando....")

#Ciclo for para separar a los números primos de los demás
print("Estos son los números primos del 1 al 100")

for primos in range(1,100): 
    if primos>1:
        for x in range(2,primos):
            if(primos % x==0): 
                break
        else: 
            print("Número: ",primos) 
#Finalizado el ciclo for se termina el programa con una frase de despedida y agradecimiento 
print("Gracias por usar el programa, hasta pronto!!!")
```
